```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading data
X = pd.read_csv("./Training Data hardwork challenge/Linear_X_Train.csv")
y = pd.read_csv("./Training Data hardwork challenge/Linear_Y_Train.csv")
X.shape
```

```python
# normalization 
X = np.array(X)
y = np.array(y)
mean = X.mean()
sv = X.std()
X = (X-mean)/sv
plt.style.use("seaborn")
plt.scatter(X,y)
plt.show()
```

```python
def hypothesis(X, theta):
    return theta[0] + theta[1]*X
```

```python
def gradient(X, Y, theta):
    m = X.shape[0]
    grad = np.zeros((2,))
    for i in range(m):
        x = X[i]
        y_ = hypothesis(x, theta)
        y = Y[i]
        grad[0] += (y_ - y)
        grad[1] += (y_ - y)*x
    return grad/m
```

```python
def gradientDescent(X, Y , max_step = 100, learning_rate = 0.00999999):
    theta = np.zeros((2,))
    grad = gradient(X, Y, theta)
    for i in range(max_step):
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]
    return theta

def error(X, Y, theta):
    m = X.shape[0]
    e = 0.0
    for i in range(m):
        y_ = hypothesis(X[i], theta)
        e += (y_ - Y[i])
    return e/m
```

```python
theta = gradientDescent(X, y)
y_ = hypothesis(X, theta) + error(X, y, theta)
```

```python
plt.scatter(X, y)
plt.plot(X, y_, color = 'orange', label = 'Predictions')
plt.legend()
plt.show()

# Accuracy of model
def r2_score(Y, Y_):
    num = np.sum((Y-Y_)**2)
    denom = np.sum((Y-Y.mean())**2)
    score = (1 - num/denom)
    return score*100
print(r2_score(y, y_))
```

```python
# result
X_Test = pd.read_csv("./Test Cases hardwork challenge/Linear_X_Test.csv")
Y_Test = hypothesis(X_Test, theta)
print(Y_Test)
```

```python
# saving in a csv file
df = pd.DataFrame(Y_Test)
df.columns = ['y']
df.to_csv("y_predictions.csv", index = False)
```
